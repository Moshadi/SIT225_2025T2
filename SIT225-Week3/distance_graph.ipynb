```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file (adjust the path if needed)
df = pd.read_csv("ultrasonic sensor project-ultrasonicDistance.csv")

# Convert 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Plot distance over time
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['value'], marker='o', linestyle='-', label='Distance (cm)', color='blue')

# Graph formatting
plt.title("Ultrasonic Distance Over Time")
plt.xlabel("Time")
plt.ylabel("Distance (cm)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
